As we found out there is only one instrument type and only one version in the whole dataset


So we can delete the instrument and version column from the dataset

```python
df1 = df.drop(['instrument','version','track'], axis='columns')
df1
```

we also found out that the acq_date column is in string format and we need to convert it to date format so as to perform different manupulations on it

```python
df1.acq_date = pd.to_datetime(df1.acq_date)
df1
```

Conerting the values of daynight and satelliete column to 0 and 1 as there are only 2 data entries..

```python
daynight_map = {"D": 1, "N": 0}
satellite_map = {"Terra": 1, "Aqua": 0}

df1.daynight = df1.daynight.map(daynight_map)
df1.satellite = df1.satellite.map(satellite_map)

df1
```

As we found out the type column has only values 0, 2 and 3 

So we are going to use dummy variable column to replace it

```python
types = pd.get_dummies(df1.type)
types
```

```python
df2 = df1.drop(['type'],axis=1)
df2 = pd.concat([df2,types], axis = 1)
df2
```

```python
df2 = df2.rename(columns={0: 'type_0', 2: 'type_2', 3: 'type_3'})
df2
```

The range of scan column is from 1 to 4.8

So smoothing the values of the column by binning method

```python
bins = [0, 1, 2, 3, 4, 5]
labels = [1,2,3,4,5]
df2['scan_binned'] = pd.cut(df2['scan'], bins=bins, labels=labels)
df3 = df2.drop(['scan'],axis = 1)
df3
```

Making a year column with the help of acq_date column

```python
df3['year'] = df3.acq_date.dt.year
df3['month'] = df3.acq_date.dt.month
df3['day'] = df3.acq_date.dt.day
df3
```

Now I will delete acq_date and acq_time column as we have divided the date column into 3 different columns

I will also delete the bright_t31 column

I also have to delete the type_0 column so as not to get in the dummy trap

```python
df4 = df3.drop(['acq_date','acq_time','bright_t31','type_0'], axis =1)
df4
```

Splitiing the Dataset into X and Y part to find the best fit model for the data

```python
x = df4.drop(['confidence'],axis = 1)
y = df4.confidence
```

Using Grid Search CV to find the best suitable model for the data

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(x,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
            }
        },
        'SVM':{
            'model':SVR(),
            'params' :{
            }
        },

        'Random_forest':{
            'model':RandomForestRegressor(),
            'params':{
            }
        },
        
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {

            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        print(algo_name)
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(x,y)
```

Now lets fine tune the parameters of Random Forest algorithm

```python
def find_best_model_using_gridsearchcv(x,y):
    algos = {
        'Random_forest':{
            'model':RandomForestRegressor(),
            'params':{
                'n_estimators' : [6,12,24,48,100,500,1000],
                'max_features' : ['sqrt', 'log2', None],
                'random_state' : [0,5,6,10,20]
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        print(algo_name)
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size =0.2, random_state = 0)

model = RandomForestRegressor(max_features = 'sqrt', n_estimators = 1000, random_state=0)

model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error

y_pred_mse = model.predict(x)

print("The mean squared error is: ", mean_squared_error(y,y_pred_mse))
```
